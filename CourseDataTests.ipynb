```python
%%sql
SELECT DISTINCT (primary_instructor) AS Teacher, COUNT(section) AS No_Sections
FROM COURSE_OFFERING
JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
WHERE act > 10 AND term = 'Spring2019'
GROUP BY primary_instructor
ORDER BY No_sections DESC
LIMIT 10;
```

```python
%%sql
SELECT primary_instructor AS Teacher, title AS Class, day, start, end
FROM COURSE_OFFERING
JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
JOIN COURSE_MEETING ON (COURSE_OFFERING.course_offering_id=COURSE_MEETING.course_offering_id)
WHERE COURSE_OFFERING.CRN = '39006' AND COURSE_OFFERING.term = 'Spring2019';
```

```python
%%sql 
SELECT program_name, AVG(act) AS Average_Size
FROM COURSE_OFFERING 
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_OFFERING.CRN=COURSE_MEETING.CRN AND COURSE_OFFERING.term=COURSE_MEETING.term)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 AND location LIKE 'DSB%'
GROUP BY program_name
ORDER BY Average_Size DESC;
```

```python
%%sql
SELECT DISTINCT CRN, title AS Class, primary_instructor AS Teacher, program_name, act AS Students, term AS Semester
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Fall2018' AND program_name = 'Accounting' AND act > 10  
OR term = 'Fall2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Spring2019' AND program_name = 'Accounting' AND act > 10
OR term = 'Spring2019' AND program_name = 'Taxation' AND act > 10;
```

```python
%%sql
SELECT primary_instructor AS Teacher, COUNT(title) AS Classes
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND program_name = 'Accounting' AND act > 10
OR term = 'Summer2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Fall2018' AND program_name = 'Accounting' AND act > 10  
OR term = 'Fall2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Spring2019' AND program_name = 'Accounting' AND act > 10
OR term = 'Spring2019' AND program_name = 'Taxation' AND act > 10
OR term = 'Fall2018' AND program_name = 'Applied_Ethics' AND act > 10 AND primary_instructor = 'Joan L. Lee'
OR term = 'Spring2019' AND program_name = 'Applied_Ethics' AND act > 10 AND primary_instructor = 'Joan L. Lee'
GROUP BY primary_instructor
ORDER BY Classes DESC;
```

```python
%%sql
SELECT primary_instructor AS Teacher, COUNT(title) AS Classes, COUNT(DISTINCT title) AS Preps
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND program_name = 'Accounting' AND act > 10
OR term = 'Summer2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Fall2018' AND program_name = 'Accounting' AND act > 10  
OR term = 'Fall2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Spring2019' AND program_name = 'Accounting' AND act > 10
OR term = 'Spring2019' AND program_name = 'Taxation' AND act > 10
OR term = 'Fall2018' AND program_name = 'Applied_Ethics' AND act > 10 AND primary_instructor = 'Joan L. Lee'
OR term = 'Spring2019' AND program_name = 'Applied_Ethics' AND act > 10 AND primary_instructor = 'Joan L. Lee'
GROUP BY primary_instructor
ORDER BY Classes DESC;
```

```python
%%sql 
SELECT DISTINCT COUNT(title) AS Title, program_name AS Department
FROM COURSE_OFFERING
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Spring2019' AND act > 10
OR term = 'Fall2018' AND act > 10
OR term = 'Summer2018' AND act > 10
GROUP BY program_name;
```

```python
%%sql 
SELECT program_name AS Department, COUNT(DISTINCT location) AS Location
FROM COURSE_OFFERING
    JOIN CATALOG_DESC ON (CATALOG_DESC.catalog_desc_id=COURSE_OFFERING.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_MEETING.course_offering_id=COURSE_OFFERING.course_offering_id)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 AND location LIKE 'DSB%'
OR COURSE_OFFERING.term = 'Fall2018' AND act > 10 AND location LIKE 'DSB%'
OR COURSE_OFFERING.term = 'Summer2018' AND act > 10 AND location LIKE 'DSB%'
GROUP BY program_name;
```

```python
%%sql 
SELECT program_name AS Department, COUNT(DISTINCT location) AS Location
FROM COURSE_OFFERING
    JOIN CATALOG_DESC ON (CATALOG_DESC.catalog_desc_id=COURSE_OFFERING.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_MEETING.course_offering_id=COURSE_OFFERING.course_offering_id)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 AND location LIKE 'DSB%'
GROUP BY program_name
ORDER BY Location DESC;
```

```python
%%sql
SELECT progam_name AS Department, COUNT(program_name), location
FROM COURSE_OFFERING
    JOIN CATALOG_DESC ON (CATALOG_DESC.catalog_desc_id=COURSE_OFFERING.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_MEETING.course_offering_id=COURSE_OFFERING.course_offering_id)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 AND location LIKE 'DSB%'
GROUP BY location;
```

```python
%%sql 
SELECT COURSE_OFFERING.term, AVG(act) AS Average_Size_DSB_Class
FROM COURSE_OFFERING 
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_OFFERING.CRN=COURSE_MEETING.CRN AND COURSE_OFFERING.term=COURSE_MEETING.term)
WHERE act > 10 AND location LIKE 'DSB%'
GROUP BY COURSE_OFFERING.term;
```

```python
%%sql
SELECT INSTRUCTOR.instructor_id, primary_instructor AS Teacher, COUNT(title) AS Classes, COUNT(DISTINCT title) AS Preps, program_name AS Department
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND act > 10
OR term = 'Fall2018' AND act > 10
OR term = 'Spring2019' AND act > 10
GROUP BY primary_instructor
ORDER BY Classes DESC
LIMIT 10;
```
