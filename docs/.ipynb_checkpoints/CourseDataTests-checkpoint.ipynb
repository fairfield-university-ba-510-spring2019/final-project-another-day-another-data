```python
%%sql

/* Printing out the individual class meetings for our Database Class */
SELECT primary_instructor AS Teacher, title AS Class, day, start, end
FROM COURSE_OFFERING
JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
JOIN COURSE_MEETING ON (COURSE_OFFERING.course_offering_id=COURSE_MEETING.course_offering_id)
WHERE COURSE_OFFERING.CRN = '39006' AND COURSE_OFFERING.term = 'Spring2019';
```

```python
%%sql

/* We are checking to see who taught the most classes in Spring2019 */
SELECT DISTINCT (primary_instructor) AS Teacher, COUNT(section) AS No_Sections
FROM COURSE_OFFERING
JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
WHERE act > 10 AND term = 'Spring2019'
GROUP BY primary_instructor
ORDER BY No_sections DESC
LIMIT 10;
```

```python
%%sql 

/* Calculating the average class size for every Department in the school for Spring 2019 */
SELECT program_name AS Department, AVG(act) AS Average_Size
FROM COURSE_OFFERING 
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_OFFERING.CRN=COURSE_MEETING.CRN AND COURSE_OFFERING.term=COURSE_MEETING.term)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 
GROUP BY Department
ORDER BY Average_Size DESC
LIMIT 10;
```

```python
%%sql 
/* Printing out average class size of DSB by last six semesters */
SELECT COURSE_OFFERING.term, AVG(act) AS Average_Size_DSB_Class
FROM COURSE_OFFERING 
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_OFFERING.CRN=COURSE_MEETING.CRN AND COURSE_OFFERING.term=COURSE_MEETING.term)
WHERE act > 10 AND location LIKE 'DSB%'
GROUP BY COURSE_OFFERING.term;
```

```python
%%sql 

/* Calculating the average class size for every Department in the business school */
SELECT program_name AS Department, AVG(act) AS Average_Size
FROM COURSE_OFFERING 
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
    JOIN COURSE_MEETING ON (COURSE_OFFERING.CRN=COURSE_MEETING.CRN AND COURSE_OFFERING.term=COURSE_MEETING.term)
WHERE COURSE_OFFERING.term = 'Spring2019' AND act > 10 AND location LIKE "%DSB%"
GROUP BY Department
ORDER BY Average_Size DESC
LIMIT 10;
```

```python
%%sql

/* Printed out every Accounting class that was offered in the last year (Summer 2018, Fall 2018, and Spring 2019) */
SELECT DISTINCT CRN, title AS Class, primary_instructor AS Teacher, program_name, act AS Students, term AS Semester
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND program_name = 'Accounting' AND act > 10 
OR term = 'Summer2018' AND program_name = 'Taxation' AND act > 10 
OR term = 'Fall2018' AND program_name = 'Accounting' AND act > 10  
OR term = 'Fall2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Spring2019' AND program_name = 'Accounting' AND act > 10
OR term = 'Spring2019' AND program_name = 'Taxation' AND act > 10;
```

```python
%%sql

/* Creating a list of the number of classes a teacher has taught and the number of distinct preps they have to done
since 2014*/
SELECT INSTRUCTOR.instructor_id, primary_instructor, COUNT(title) AS Classes_Taught, COUNT(DISTINCT title) AS Distinct_Preps
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
WHERE act > 10
GROUP BY primary_instructor
ORDER BY Classes_Taught DESC
LIMIT 10;
```

```python
%%sql

/* Creating a list of the number of classes a teacher has taught and the number of distinct preps they have to done
in the last year*/
SELECT INSTRUCTOR.instructor_id, primary_instructor, COUNT(title) AS Classes_Taught, COUNT(DISTINCT title) AS Distinct_Preps, program_name 
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND act > 10
OR term = 'Fall2018' AND act > 10
OR term = 'Spring2019' AND act > 10
GROUP BY primary_instructor
ORDER BY Classes_Taught DESC
LIMIT 10;
```

```python
%%sql

/* Creating a list of the number of classes an Accounting teacher has taught and the number of distinct 
preps they have to done in the last year*/
SELECT primary_instructor AS Teacher, COUNT(title) AS Classes, COUNT(DISTINCT title) AS Preps
FROM COURSE_OFFERING
    JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id)
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Summer2018' AND program_name = 'Accounting' AND act > 10
OR term = 'Summer2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Fall2018' AND program_name = 'Accounting' AND act > 10  
OR term = 'Fall2018' AND program_name = 'Taxation' AND act > 10
OR term = 'Spring2019' AND program_name = 'Accounting' AND act > 10
OR term = 'Spring2019' AND program_name = 'Taxation' AND act > 10
GROUP BY primary_instructor
ORDER BY Classes DESC;
```

```python
%%sql 
/* Creating a list of the different number of classes a department offered in the last year */
SELECT DISTINCT COUNT(title) AS Title, program_name AS Department
FROM COURSE_OFFERING
    JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id)
WHERE term = 'Spring2019' AND act > 10
OR term = 'Fall2018' AND act > 10
OR term = 'Summer2018' AND act > 10
GROUP BY program_name
ORDER BY Title DESC
LIMIT 10;
```

```python
#Imported packages in order to be able to pull a table to a dataframe
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine("sqlite:///../FP/CourseData-Copy1.db")
```

```python
#Used an engine to pull select teacher information to a dataframe that will be used in our datawarehouse creation 
with engine.connect() as conn, conn.begin():
    Teacher_Info = pd.read_sql("SELECT INSTRUCTOR.instructor_id, primary_instructor, COUNT(title) AS classes_taught, COUNT(DISTINCT title) AS distinct_preps, program_name FROM COURSE_OFFERING JOIN INSTRUCTOR ON (COURSE_OFFERING.instructor_id=INSTRUCTOR.instructor_id) JOIN CATALOG_DESC ON (COURSE_OFFERING.catalog_desc_id=CATALOG_DESC.catalog_desc_id) WHERE term = 'Summer2018' AND act > 10 OR term = 'Fall2018' AND act > 10 OR term = 'Spring2019' AND act > 10 GROUP BY primary_instructor ORDER BY Classes_Taught DESC;", conn)                                
```

```python
#Saved the Teacher_Info dataframe to a csv so can be imported into our datawarehouse
Teacher_Info.to_csv('teacher_info.csv', index=False)
```

```python
#Used an engine to pull select teacher information to a dataframe that will be used in our datawarehouse creation 
with engine.connect() as conn, conn.begin():
    course_info = pd.read_sql("SELECT COURSE_OFFERING.CRN, instructor_id, catalog_desc_id, COURSE_OFFERING.course_offering_id, location, cap, act, rem FROM COURSE_OFFERING JOIN COURSE_MEETING ON (COURSE_MEETING.course_offering_id=COURSE_OFFERING.course_offering_id) WHERE COURSE_OFFERING.term = 'Summer2018' AND act > 10 OR COURSE_OFFERING.term = 'Fall2018' AND act > 10 OR COURSE_OFFERING.term = 'Spring2019' AND act > 10;", conn)
```

```python
#Saved the course_Info dataframe to a csv so can be imported into our datawarehouse
course_info.to_csv('course_info.csv', index=False)
```
